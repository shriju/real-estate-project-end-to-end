```python
import numpy as np
import pandas as pd
```

```python
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
```

```python
flats= pd.read_csv("/content/flats_cleaned (2).csv")
houses = pd.read_csv("/content/house_cleaned (2).csv")
```

```python
df=pd.concat([flats, houses], ignore_index=True)
```

```python
df= df.sample(df.shape[0],ignore_index=True)
```
